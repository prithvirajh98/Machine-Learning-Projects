## <span style = 'color:Indigo'> DATA PRE-PROCESSING </span>
---

### <span style = 'color:Green'> <i> Seperate the Class Feature for further Analysis </i> </span>
---

  * All the transactions which are legitimate will be stored in variable called **"legitimate_Transactions"**.
  

  * All the transactions which are fradulant will be stored in variable called ***"Fradulent_Transactions"**.

```python
legitimate_Transactions = creditCard_df[creditCard_df.Class == 0]
Fraudulent_Transactions = creditCard_df[creditCard_df.Class == 1]
```

```python
print(f"The shape of the Legitimate Transactions : {legitimate_Transactions.shape}")
print()
print(f"The shape of the Fradulent Transactions : {Fraudulent_Transactions.shape}")
```

#### <span style = 'color:Orange'> Under-Sampling </span>
---

* Since the data contains more than 90% of the legit transactions, it is necessary to build a sample dataset that contains similar distribution of Legit transactions and Fraudulent Transactions.


* Extract 492 random legit transactions and append it to the fraudulent transactions.


* This makes the distribution uniform and the makes it a better dataset for the predictive model.

```python
legitSamples = legitimate_Transactions.sample(n=492)
```

#### <span style= 'color:Orange'> Uniform Data: Concatinate the Fraud_Transactions and legit_Sample in order to make the distribution of the dataset uniform </span>
---

```python
creditCardData = pd.concat([legitSamples, Fraudulent_Transactions], axis = 0)
```

#### <span style = 'color:Orange'> Initial rows of the Normalized Data </span>
---

```python
print(f"Initial rows of the Normalised Data:")
creditCardData.head()
```

#### <span style = 'color:Orange'> Final rows of the Normalized Data </span>
---

```python
print(f"Initial rows of the Normalised Data:")
creditCardData.tail()
```

#### <span style= 'color:Orange'> Number of Legimate Transactions and Fradualent Transactions of the new dataset </span>
---

```python
creditCardData.Class.value_counts()
```

  * <span style = 'color:Red'> **Now the data is uniformly distributed where there are 492 Legitimate and Fraudulant Transactions.** </span>

#### <span style= 'color:Orange'> Compare the average of both categories of transactions.</span>
---

```python
creditCardData.groupby('Class').mean()
```

---

## <span style = 'color:Indigo'> SEGREGATE THE DATA INTO DEPENDENT AND INDEPENDET FEATURES </span>
---

```python
X = creditCardData.drop(columns = 'Class', axis = 1)
y = creditCardData['Class']
```

---

## <span style = 'color:Indigo'> SPLIT THE DATA INTO TRAINING AND TESTING DATASET </span>
---

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify = y, random_state = 2)
```

  * <span style = 'color:Red'>**Startify is used to keep the proportion of y values through the training and test sets.**</span>

---

## <span style = 'color:Indigo'> BUILDING THE MODEL </span>
---

### <span style = 'color:Orange'> The Logistics Regression Model </span>
---

   * It is the a machine learning model that comes under **Supervised Learning Technique.**
   
   
   * It is used to predict the categorical dependent variables using a given set of independent variables.

```python
model = LogisticRegression()
```

### <span style= 'color:Orange'> Training the Logistic Regression Model using the Training Dataset </span>
---

```python
model.fit(X_train, y_train)
```

### <span style= 'color:Orange'> Model Evaluation. </span>
---

  * Accuracy Score --> Training Data

```python
X_train_prediction = model.predict(X_train)

print(f"The accuracy of the model with training data : {round((accuracy_score(X_train_prediction, y_train) * 100), 2)}%")
```

  * Accuracy Score --> Testing Data

```python
X_test_Prediction = model.predict(X_test)

print(f"The accuracy of the model with testing data : {round((accuracy_score(X_test_Prediction, y_test) * 100), 2)}%")
```
